Because of 115296 null values in CHQ.NO.out of 116201 rows,we are going to drop this column.

```python
#Dropping CHQ.NO. column
df.drop('CHQ.NO.',axis=1,inplace=True)
```

```python
#Now,we are going to the fill the null values with median of the respective columns
df['DEPOSIT AMT'].fillna(df['DEPOSIT AMT'].median(),inplace=True)
df['WITHDRAWAL AMT'].fillna(df['WITHDRAWAL AMT'].median(),inplace=True)
```
